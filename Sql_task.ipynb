```python
import sqlite3

# Create database connection
conn = sqlite3.connect("code_review.db")
cursor = conn.cursor()

# Create table
cursor.execute("""
CREATE TABLE IF NOT EXISTS code_reviews (
    review_id INTEGER PRIMARY KEY AUTOINCREMENT,
    user_id INTEGER,
    language TEXT,
    code TEXT,
    feedback TEXT,
    review_date TIMESTAMP DEFAULT CURRENT_TIMESTAMP
);
""")

conn.commit()
conn.close()

print("Table created successfully!")
```

```python
import sqlite3

# 🔹 Connect to database
def connect_db():
    return sqlite3.connect("code_review.db")


# 🔹 Create table
def create_table():
    conn = connect_db()
    cursor = conn.cursor()

    cursor.execute("""
    CREATE TABLE IF NOT EXISTS code_reviews (
        review_id INTEGER PRIMARY KEY AUTOINCREMENT,
        user_id INTEGER,
        language TEXT,
        code TEXT,
        feedback TEXT,
        review_date TIMESTAMP DEFAULT CURRENT_TIMESTAMP
    )
    """)

    conn.commit()
    conn.close()
    print("Table ready ✅")


# 🔹 Insert data
def insert_review(user_id, language, code, feedback):
    conn = connect_db()
    cursor = conn.cursor()

    cursor.execute("""
    INSERT INTO code_reviews (user_id, language, code, feedback)
    VALUES (?, ?, ?, ?)
    """, (user_id, language, code, feedback))

    conn.commit()
    conn.close()
    print("Review inserted ✅")


# 🔹 View all reviews
def view_reviews():
    conn = connect_db()
    cursor = conn.cursor()

    cursor.execute("SELECT * FROM code_reviews")
    rows = cursor.fetchall()

    for row in rows:
        print(row)

    conn.close()


# 🔹 Run functions
create_table()
insert_review(101, "Python", "print('Hello')", "Good code structure")
view_reviews()
```
